```python
def calculate_score(candidate):

    # Skills
    jd_skills = [
        "NLP",
        "Fine-tuning LLMs",
        "LoRA",
        "Milvus",
        "BentoML"
    ]

    candidate_skill_names = [
        skill["name"]
        for skill in candidate["skills"]
    ]

    skill_score = 0

    for skill in candidate_skill_names:
        if skill in jd_skills:
            skill_score += 1

    # Experience

    years_exp = candidate["profile"]["years_of_experience"]

    if 5 <= years_exp <= 9:
        experience_score = 10
    elif 3 <= years_exp < 5:
        experience_score = 7
    else:
        experience_score = 4

    # Behavior

    behavior_score = 0

    signals = candidate["redrob_signals"]

    if signals["open_to_work_flag"]:
        behavior_score += 2

    if signals["interview_completion_rate"] > 0.7:
        behavior_score += 2

    if signals["recruiter_response_rate"] > 0.3:
        behavior_score += 2

    if signals["github_activity_score"] > 0:
        behavior_score += 2

    if signals["profile_completeness_score"] > 80:
        behavior_score += 2

    final_score = (
        skill_score +
        experience_score +
        behavior_score
    )

    return final_score
```

```python
def calculate_experience_score(candidate):

    years = candidate["profile"]["years_of_experience"]

    if 5 <= years <= 9:
        return 10

    elif 4 <= years < 5:
        return 8

    elif 9 < years <= 12:
        return 7

    else:
        return 4
```

```python
def calculate_behavior_score(candidate):

    signals = candidate["redrob_signals"]

    score = 0

    if signals["open_to_work_flag"]:
        score += 2

    if signals["interview_completion_rate"] > 0.7:
        score += 2

    if signals["recruiter_response_rate"] > 0.3:
        score += 2

    if signals["profile_completeness_score"] > 80:
        score += 2

    if signals["github_activity_score"] > 0:
        score += 2

    return score
```

```python
def calculate_score_v2(candidate):

    skill_score = calculate_skill_score(candidate)

    experience_score = calculate_experience_score(candidate)

    behavior_score = calculate_behavior_score(candidate)

    ai_score = calculate_ai_relevance(candidate)

    final_score = (
        skill_score * 0.30 +
        experience_score * 0.20 +
        behavior_score * 0.20 +
        ai_score * 0.30
    )

    return round(final_score, 2)
```

```python
def calculate_ai_relevance(candidate):

    text = ""

    text += candidate["profile"]["headline"] + " "
    text += candidate["profile"]["summary"] + " "

    for job in candidate["career_history"]:
        text += job["title"] + " "
        text += job["description"] + " "

    text = text.lower()

    weights = {
        "retrieval": 5,
        "ranking": 5,
        "embedding": 5,
        "recommendation": 4,
        "machine learning": 3,
        "nlp": 3,
        "llm": 3,
        "vector": 3,
        "search": 2,
        "fine-tuning": 2,
        "rag": 2
    }

    score = 0

    for keyword, weight in weights.items():
        if keyword in text:
            score += weight

    return score
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
def build_candidate_text(candidate):

    text_parts = []

    profile = candidate["profile"]

    text_parts.append(profile.get("headline", ""))
    text_parts.append(profile.get("summary", ""))

    for skill in candidate["skills"]:
        text_parts.append(skill["name"])

    for job in candidate["career_history"]:
        text_parts.append(job.get("title", ""))
        text_parts.append(job.get("description", ""))

    return " ".join(text_parts)
```

```python
jd_text = """
Senior AI Engineer

Must have:
Embeddings
Retrieval Systems
Ranking Systems
Vector Databases
Machine Learning
Python
Recommendation Systems
Search
NLP
LLMs
Production ML

Experience building scalable AI products.

Strong understanding of NDCG, MAP, MRR evaluation metrics.

Experience with retrieval and ranking systems.
"""
```

```python
candidate31 = None

for candidate in candidates:
    
    if candidate["candidate_id"] == "CAND_0000031":
        
        candidate31 = candidate
        break
```

```python
candidate1_text = build_candidate_text(candidates[0])

candidate31_text = build_candidate_text(candidate31)
```

```python
jd_embedding = model.encode([jd_text])

candidate1_embedding = model.encode([candidate1_text])

candidate31_embedding = model.encode([candidate31_text])
```

```python
score1 = cosine_similarity(
    jd_embedding,
    candidate1_embedding
)[0][0]

score31 = cosine_similarity(
    jd_embedding,
    candidate31_embedding
)[0][0]

print("Candidate 1:", round(score1,4))
print("Candidate 31:", round(score31,4))
```

```python
def calculate_semantic_score(candidate):

    candidate_text = build_candidate_text(candidate)

    candidate_embedding = model.encode(
        [candidate_text]
    )

    score = cosine_similarity(
        jd_embedding,
        candidate_embedding
    )[0][0]

    return float(score)
```

```python
def calculate_score_v4(candidate):

    semantic_score = calculate_semantic_score(candidate)

    experience_score = (
        calculate_experience_score(candidate) / 10
    )

    behavior_score = (
        calculate_behavior_score(candidate) / 10
    )

    final_score = (
        0.70 * semantic_score +
        0.15 * experience_score +
        0.15 * behavior_score
    )

    return round(final_score,4)
```

```python
scores = []

for candidate in candidates:

    scores.append({
        "candidate_id": candidate["candidate_id"],
        "score": calculate_score_v4(candidate)
    })

scores_df = pd.DataFrame(scores)

scores_df = scores_df.sort_values(
    by="score",
    ascending=False
)

scores_df.head(10)
```

```python
def passes_ai_filter(candidate):

    text = build_candidate_text(candidate).lower()

    strong_signals = [
        "machine learning",
        "ml engineer",
        "data scientist",
        "retrieval",
        "ranking",
        "embedding",
        "recommendation",
        "nlp",
        "llm",
        "deep learning",
        "vector",
        "fine-tuning",
        "rag"
    ]

    matches = 0

    for signal in strong_signals:
        if signal in text:
            matches += 1

    return matches >= 2
```
