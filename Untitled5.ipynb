```python
from tqdm import tqdm

candidate_profiles = []

for c in tqdm(candidates):

    profile = c.get("profile", {})
    career = c.get("career_history", [])
    skills = c.get("skills", [])

    headline = profile.get("headline", "")
    summary = profile.get("summary", "")
    current_title = profile.get("current_title", "")

    skill_text = " ".join([
        skill.get("name", "")
        for skill in skills
    ])

    career_text = " ".join([
        job.get("description", "")
        for job in career
    ])

    combined_profile = f"""
    Current Role: {current_title}
    Headline: {headline}

    Summary:
    {summary}

    Skills:
    {skill_text}

    Career History:
    {career_text}
    """

    candidate_profiles.append(combined_profile)

print("Profiles Created:", len(candidate_profiles))
```

```python
hybrid_features = []

for idx, candidate in enumerate(candidates):

    signals = candidate.get("redrob_signals", {})
    profile = candidate.get("profile", {})

    semantic_score = semantic_scores[idx]

    profile_score = signals.get(
        "profile_completeness_score",
        0
    ) / 100

    recruiter_response = signals.get(
        "recruiter_response_rate",
        0
    )

    github_score = signals.get(
        "github_activity_score",
        0
    ) / 100

    saved_by_recruiters = min(
        signals.get("saved_by_recruiters_30d", 0) / 25,
        1
    )

    open_to_work = (
        1 if signals.get("open_to_work_flag", False)
        else 0
    )

    years_exp = profile.get(
        "years_of_experience",
        0
    )

    exp_score = max(
        0,
        1 - abs(years_exp - 7) / 7
    )

    hybrid_features.append({
        "candidate_id": candidate["candidate_id"],
        "semantic_score": semantic_score,
        "profile_score": profile_score,
        "recruiter_response": recruiter_response,
        "github_score": github_score,
        "saved_by_recruiters": saved_by_recruiters,
        "open_to_work": open_to_work,
        "experience_score": exp_score
    })
```

```python
hybrid_df = pd.DataFrame(
    hybrid_features
)
```

```python
hybrid_df["final_score"] = (
    0.60 * hybrid_df["semantic_score"] +
    0.10 * hybrid_df["profile_score"] +
    0.10 * hybrid_df["recruiter_response"] +
    0.05 * hybrid_df["github_score"] +
    0.05 * hybrid_df["saved_by_recruiters"] +
    0.05 * hybrid_df["open_to_work"] +
    0.05 * hybrid_df["experience_score"]
)
```

```python
final_rankings = hybrid_df.sort_values(
    "final_score",
    ascending=False
)

final_rankings.head(20)
```

```python
career_keywords = [
    "search",
    "retrieval",
    "recommendation",
    "ranking",
    "ml engineer",
    "ai engineer",
    "applied ml",
    "machine learning",
    "nlp"
]

research_keywords = [
    "research scientist",
    "research engineer",
    "academic",
    "phd",
    "research lab"
]
```

```python
career_scores = []

for candidate in candidates:

    text = candidate_profiles[
        candidates.index(candidate)
    ].lower()

    positive = sum(
        keyword in text
        for keyword in career_keywords
    )

    negative = sum(
        keyword in text
        for keyword in research_keywords
    )

    score = max(
        0,
        (positive * 0.15) - (negative * 0.10)
    )

    career_scores.append(score)
```

```python
hybrid_df["career_score"] = career_scores
```

```python
hybrid_df["final_score_v2"] = (
    0.55 * hybrid_df["semantic_score"] +
    0.15 * hybrid_df["career_score"] +
    0.10 * hybrid_df["recruiter_response"] +
    0.05 * hybrid_df["profile_score"] +
    0.05 * hybrid_df["github_score"] +
    0.05 * hybrid_df["saved_by_recruiters"] +
    0.05 * hybrid_df["experience_score"]
)
```

```python
def generate_smart_reason(candidate):

    profile = candidate["profile"]
    career = candidate["career_history"]

    title = profile.get("current_title", "")
    exp = profile.get("years_of_experience", 0)

    evidence = []

    for job in career:

        text = job.get("description", "").lower()

        if "ranking" in text:
            evidence.append("ranking systems")

        if "retrieval" in text:
            evidence.append("retrieval systems")

        if "search" in text:
            evidence.append("search infrastructure")

        if "embedding" in text:
            evidence.append("embedding-based solutions")

        if "recommendation" in text:
            evidence.append("recommendation systems")

    evidence = list(set(evidence))

    evidence_text = ", ".join(evidence[:3])

    return (
        f"{title} with {exp:.1f} years of experience. "
        f"Demonstrated production experience in "
        f"{evidence_text}, closely matching the Senior AI Engineer "
        f"requirements around retrieval, ranking and applied ML systems."
    )
```

```python
reasons = []

for cid in top100["candidate_id"]:

    reasons.append(
        generate_smart_reason(
            candidate_lookup[cid]
        )
    )

submission_df["reasoning"] = reasons
```
